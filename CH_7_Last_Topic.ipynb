```python
import requests
url='https://jsonplaceholder.typicode.com/posts'
page=1
while page<=3:
    parameter={'_page':page,'_limit':5}
    response=requests.get(url,params=parameter)
    data=response.json()
    print("Post on page",page)
    for i in data:
        print(i['id'],":",i["title"])
    page+=1
```
