```python
import requests
from bs4 import BeautifulSoup
website="https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
result=requests.get(website) #requests send karie che,data leva mate (object formate ma data aavshe)
content=result.text #object to text
# content # output string ma aavshe
soup=BeautifulSoup(content,"html.parser") #string to html 
# soup
# print(soup.prettify)
```

```python
name=[]
price=[]
detail=soup.find_all('div',class_='card thumbnail')
for i in detail:
    name.append(i.find('a').text.strip())
    price.append(i.find('span').text.strip())
data={"leptop_name : ":name,"Leptop_price : ":price}
import pandas as pd
df=pd.DataFrame(data)
df.to_csv('Leptop.csv')
```

# 30 most popular movie

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
with open("30 Most Popular Movies.html","r",encoding="utf-8") as file:
    content=file.read()
 
soup=BeautifulSoup(content,'html.parser')
box=soup.find_all('div',class_='article_movie_title')
Film_name=[]
rating=[]
for i1 in box:
    Film_name.append(i1.find('a').text.strip())
    rating.append(i1.find('span',class_='tMeterScore').text.strip())

data={"Film_Name : ":Film_name,"Film_Rating : ":rating}
df=pd.DataFrame(data)
df.to_csv('A30_movie.csv',index=False)
df  
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
with open('Certified used Mercedes-Benz for sale.html','r',encoding='utf-8')as file:
    content=file.read()

soup=BeautifulSoup(content,'html.parser')
box=soup.find_all('div',class_='vehicle-details')
name=[]
milage=[]
price=[]
rat=[]
for i in box:
    name.append(i.find('h2',class_='title').text.strip())
    milage.append(i.find('div',class_='mileage').text.strip())
    price.append(i.find('span',class_='primary-price').text.strip())
    # rat.append(i.find('span',class_='sds-rating__count').text.strip())
    r=i.find('span',class_='sds-rating__count')

    if r:
        rat.append(r.text.strip())

    else:
        rat.append("not available")


data={"Name : ":name,"Milage : ":milage,"Prie : ":price,"Rating : ":rat}  
df=pd.DataFrame(data)
df.to_csv('car.csv',index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
with open('flipkart.html','r',encoding='utf-8')as file:
    content=file.read()
name=[]
rating=[]
descount=[]
actual=[]
soup=BeautifulSoup(content,'html.parser')
box=soup.find_all('div',class_='col col-7-12')
for i in box:
    name.append(i.find('div',class_='_4rR01T').text.strip())
    rating.append(i.find('div',class_='_3LWZlK').text.strip())
    descount.append(i.find('div',class_='_30jeq3 _1_WHN1'))
    a=i.find('div',class_='_3I9_wc _27UcVY')
    if a:
        actual.append(a.text.strip())
    else:
         actual.append("not available")

        
        
data={"name ":name,"rating ":rating,"descount":descount,"actual ":actual}  
df=pd.DataFrame(data)
df.to_csv('flip.csv',index=False)
# name
# rating
# descount
# actual  
```
