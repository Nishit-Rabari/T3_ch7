```python
website="https://subslikescript.com/movies"
result=requests.get(website) #requests send karie che,data leva mate (object formate ma data aavshe)
content=result.text #object to text
# content # output string ma aavshe
soup=BeautifulSoup(content,"html.parser") #string to html 
soup
print(soup.prettify)
```

```python
box=soup.find("article",class_="main-article")
movie_list=box.find_all("a")# fin_all answer list ma return kare...
movie_title=[]
for i in movie_list:
    movie_title.append(i.text.strip())
    
movie_title    
```

```python
links=[]
for i in box.find_all('a',href=True):
    links.append(i['href'])
links
```

```python
root="https://subslikescript.com"
for i in links:
    website=f"{root}/{i}"
    result=requests.get(website)
    content=result.text
    soup=BeautifulSoup(content,"html.parser") #string to html 
    transcript=soup.find("div",class_='full-script').text
    T=soup.find("h1").text.strip()
    with open(T+".txt","w",encoding="utf-8") as file:
        file.write(transcript)
    print("complate")
    

```
