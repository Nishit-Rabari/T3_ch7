# send Quary Paramiter

```python
param={'userId':1}
url='https://jsonplaceholder.typicode.com/posts'
response=requests.get(url,params=param)
filter_post=response.json()
print(len(filter_post))
print(json.dumps(filter_post,indent=5))
```

# custom method

```python
# 1. send post requests
url='https://jsonplaceholder.typicode.com/posts'
new_post={'title':'ljiet','body':'this is collage','userId':1}
response=requests.post(url,json=new_post)
print(response.status_code)
print(response.json()) # actual webpage per dekhashe nahi
```

```python
# 2. send put requests (update karva mate use thay...)(post na all parameter change karva pade)
url='https://jsonplaceholder.typicode.com/posts/1'
updated_post={'id':1,'title':'updated_title','body':'updated_body','userId':1} # first data ne update karva mate /1 lakhvanu
response=requests.put(url,json=updated_post)
print(response.status_code)
print(response.json())
```

```python
#3.send patch requests (koi post na koi 1 parameter change kari shakay)
url='https://jsonplaceholder.typicode.com/posts/1'
patch_data={'title':'LJIET'}
response=requests.patch(url,json=patch_data)
print(response.status_code)
print(response.json())
```

```python
# 4. send Delete requests
url='https://jsonplaceholder.typicode.com/posts/1'
delete=requests.delete(url)
print(delete.status_code)
```

```python
# 5. view response header
response=requests.get(url)
print(response.headers)
```
